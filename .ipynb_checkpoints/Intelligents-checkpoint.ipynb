## Mudeli tiendamine

### Data augmentation - andmete rikastamine

Mudelit kauem treenides hakkab mudel lesobitama, kuna pib ra tundma just neid kindlaid pilte. Selleks, et mudel pigem edukalt ldistaks ja ei keskenduks kindlatele piltidele kasutame andmete rikastamist. Selleks prame ja peegeldame pilti, suurendame, lisame "soola" jne.

#### Vaatame treeningandmeid

```python
def get_augs():
    tfms = tfms_from_model(resnet34, sz, aug_tfms=transforms_side_on, max_zoom=1.1)
    data = ImageClassifierData.from_paths(PATH, bs=2, tfms=tfms, num_workers=1)
    x,_ = next(iter(data.aug_dl))
    return data.trn_ds.denorm(x)[1]
ims = np.stack([get_augs() for i in range(6)])
plots(ims, rows=2)
```

```python
bs=10
```

```python
def get_data(sz, bs):
    tfms = tfms_from_model(resnet34, sz, aug_tfms=transforms_side_on, max_zoom=1.1)
    data = ImageClassifierData.from_paths(PATH, bs=bs, tfms=tfms, num_workers=1,test_name='test')
    
    return data
```

```python
data = get_data(sz, bs)
```

```python
learn = ConvLearner.pretrained(arch, data, precompute=True)
```

```python
learn.fit(1e-2, 3)
```

By default when we create a learner, it sets all but the last layer to *frozen*. That means that it's still only updating the weights in the last layer when we call `fit`.

```python
learn.precompute=False
```

```python
learn.fit(1e-2, 3, cycle_len=1)
```

### Tuunimine

Esialgu petasime ainult viimaseid kihte, et nende kaalud vastaksid rohkem meie piltidele. Nd avame aga treenimiseks kogu mudeli, et treenida ka esimese osa kaalusid.

```python
learn.unfreeze()
```

Mudeli esimesed kihid on juba varem treenitud Imagenet piltide peal ja me ei taha neid rikkuda.

Viimased kihid on algselt aga juhuarvudega tidetud.

Selleks, et esimesed kihid vhem mjutatud oleks kasutame mitmetasemelist pisammu. Esimeste kihtide pisamm kige viksem, kuna neid ei ole vaja eriti muuta,keskmiste kihtide pisamm vahepealne ja viimaste kihtide pisamm suurem (need kihid tunnetavad otseselt meie piltide peal olevaid objekte ja vajavad seega treenimist.

```python
lr=np.array([1e-4,1e-3,1e-2])
```

```python
learn.fit(lr, 3, cycle_len=3, cycle_mult=2)
```

```python
learn.sched.plot_lr()
```

```python
learn.sched.plot_loss()
```

```python
learn.save('224_all')
```

```python
learn.sched.plot_loss()
```

Andmete rikastamist saab edukalt kasutada ka testimise faasis. TTA - Test time augmentation.

TTA teeb ennustuse pildi kohta, aga htlasi ka mitme teise pildi kohta mis on loodud lbi andmete rikastamise (ldjuhul originaal + 4 muudetud versiooni). Seejrel vtab keskmise tulemuse ja kasutab seda.

```python
log_preds,y = learn.TTA(n_aug=4)
probs = np.mean(np.exp(log_preds),0)
```

```python
accuracy_np(probs, y)
```

## Tulemused

### Confusion matrix 

```python
preds = np.argmax(probs, axis=1)
probs = probs[:,1]
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y, preds)
```

Confusion matrix - aitab analsida mudeli tulemusi

```python
plot_confusion_matrix(cm, data.classes)
```
